```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
import matplotlib.pyplot as plt
```

```python
df_labels = pd.read_csv('C:/Users/Yannick/Google Drive/person_detection/pcds_dataset_detections/pcds_dataset_detected/pcds_dataset_labels_united.csv', names=['file_name', 'entering', 'exiting', 'video_class'])
df_labels.head()
```

```python
plt.clf()
CATEGORY_MAPPING = ['normal_uncrowded','normal_crowded','noisy_uncrowded','noisy_crowded']
ax = df_labels['video_class'].hist(label='video_class', bins = np.arange(5) - 0.25, width=0.5)
ax.set_xlabel('Video class')
ax.set_ylabel('Count for video class')
ax.set_xticks([0,1,2,3])
ax.set_xlim=[-1, 3]
ax.set_xticklabels(CATEGORY_MAPPING)
plt.savefig('C:/Users/Yannick/OneDrive/Dokumente/Studium/Masterarbeit/Ausarbeitung/98_images/histogram_pcds_classes.pgf')
```

```python
plt.clf()
plt.hist([df_labels['exiting'].values, df_labels['entering'].values], bins=max(df_labels['exiting'].max(), df_labels['exiting'].max()), label=['exiting', 'entering'], range=(0, 20))
plt.xlabel('Number of persons')
plt.ylabel('Count for number of persons')
plt.legend(loc='upper right')
plt.xticks(range(0,20,3))
plt.show()
plt.savefig('C:/Users/Yannick/OneDrive/Dokumente/Studium/Masterarbeit/Ausarbeitung/98_images/histogram_entering_exiting.pgf')
```
